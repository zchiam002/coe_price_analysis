```python
table = soup.find('table', {'class': 'table table-borderless'})
```

```python
description_identifier = '<p class="font-12 mb-0">'

count = 0 
for row in table.find_all('p'):
    print(row.text, count)
    count = count + 1
    if description_identifier in row: 
        print(row)
```
